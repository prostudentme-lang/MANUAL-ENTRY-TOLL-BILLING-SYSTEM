```python
import datetime

# Dictionary for toll rates based on vehicle type (per km)
TOLL_RATES = {
    "Car": 2,
    "Bus": 5,
    "Truck": 7,
    "Bike": 1
}

# Database to store transactions
transactions = []

def calculate_toll(vehicle_type, distance, journey_type):
    rate = TOLL_RATES.get(vehicle_type, 2)
    toll_amount = rate * distance
    if journey_type.lower() == "round-trip":
        toll_amount *= 1.8  # discounted rate for return trip
    return toll_amount

def generate_receipt(data):
    print("\n===== TOLL RECEIPT =====")
    for key, value in data.items():
        print(f"{key}: {value}")
    print("========================\n")

def main():
    while True:
        print("\n--- Toll Collection System ---")
        print("1. Collect Toll")
        print("2. View Transactions")
        print("3. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            # Collect inputs
            vehicle_number = input("Enter Vehicle Number: ")
            vehicle_type = input("Enter Vehicle Type (Car/Bus/Truck/Bike): ")
            vehicle_category = input("Enter Vehicle Category (Private/Commercial): ")
            vehicle_model = input("Enter Vehicle Model: ")
            vehicle_color = input("Enter Vehicle Color: ")
            owner_name = input("Enter Owner Name: ")
            owner_contact = input("Enter Owner Contact Number: ")
            owner_address = input("Enter Owner Address: ")
            license_no = input("Enter Driving License Number: ")
            rc_no = input("Enter RC Number: ")
            engine_no = input("Enter Engine Number: ")
            chassis_no = input("Enter Chassis Number: ")
            axles = input("Enter Number of Axles: ")
            fuel_type = input("Enter Fuel Type (Petrol/Diesel/Electric): ")
            journey_type = input("Enter Journey Type (One-way/Round-trip): ")
            entry_location = input("Enter Entry Toll Plaza Location: ")
            exit_location = input("Enter Exit Toll Plaza Location: ")
            distance = int(input("Enter Distance (km): "))
            toll_rate = TOLL_RATES.get(vehicle_type, 2)
            payment_mode = input("Enter Payment Mode (Cash/Card/UPI/Fastag): ")
            transaction_id = input("Enter Payment Transaction ID: ")
            date = datetime.date.today()
            time = datetime.datetime.now().strftime("%H:%M:%S")
            receipt_no = "REC" + str(len(transactions) + 1).zfill(5)
            operator_id = input("Enter Operator ID: ")

            # Calculate toll
            toll_amount = calculate_toll(vehicle_type, distance, journey_type)

            # Store data
            data = {
                "Receipt No": receipt_no,
                "Date": date,
                "Time": time,
                "Vehicle No": vehicle_number,
                "Vehicle Type": vehicle_type,
                "Category": vehicle_category,
                "Model": vehicle_model,
                "Color": vehicle_color,
                "Owner": owner_name,
                "Contact": owner_contact,
                "Address": owner_address,
                "License No": license_no,
                "RC No": rc_no,
                "Engine No": engine_no,
                "Chassis No": chassis_no,
                "Axles": axles,
                "Fuel": fuel_type,
                "Journey": journey_type,
                "Entry": entry_location,
                "Exit": exit_location,
                "Distance (km)": distance,
                "Rate/km": toll_rate,
                "Toll Amount": toll_amount,
                "Payment Mode": payment_mode,
                "Transaction ID": transaction_id,
                "Operator ID": operator_id
            }

            transactions.append(data)

            # Generate receipt
            generate_receipt(data)

        elif choice == "2":
            print("\n=== All Transactions ===")
            for t in transactions:
                print(t)
            print("========================")

        elif choice == "3":
            print("Exiting Toll Collection System...")
            break
        else:
            print("Invalid choice! Try again.")

if __name__ == "__main__":
    main()
```
